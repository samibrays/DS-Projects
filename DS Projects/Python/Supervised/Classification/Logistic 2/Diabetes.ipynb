# Data Modeling

```python
# split the dataset into dependant and independant
X = diabetes.drop('Diabetes_012', axis = 1)
y = diabetes['Diabetes_012']
```

```python
# check for class imbalance
y.value_counts(normalize=True)

# 0 - majority
# 1- highly inderrepresented
# 2 - minority
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# train - test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size = 0.2, random_state = 42,
    stratify = y
)
```

```python
# handle class imbalance using class weights
class_weights = {0:1 , 1:3, 2:2}

# nalizes misclassification of minority classes

# eps original data distribution
```

```python
# moedel training using multinomial logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

# Some sklearn versions may not accept the `multi_class` argument.
# Try to use multinomial; fall back to OneVsRest if not supported.
try:
    model = LogisticRegression(
        multi_class='multinomial', solver='lbfgs',
        class_weight=class_weights, max_iter=500
    )
except TypeError:
    base_clf = LogisticRegression(class_weight=class_weights, max_iter=500, solver='lbfgs')
    model = OneVsRestClassifier(base_clf)
```

```python
# model fitting
model.fit(X_train, y_train)
```

```python
# model coefficients, intercepts and interpretation
coef_df = pd.DataFrame(
    model.coef_, columns=X.columns,
    index=['Class 0', 'Class 1', 'Class 2']
)

coef_df.T.sort_values(by='Class 2', ascending=False)

# i.e , a one unit increase in BMI increases the log-odds of being in Class 2 (Diabetic) versus Class 0 (Non-Diabetic) by approximately 0.25, holding all other variables constant.
```

# Model Prediction

```python
# class predictions
y_pred = model.predict(X_test)
```

```python
# probability predictions
y_prob = model.predict_proba(X_test)
```

```python
# actual vs predicted vs prbobality comparison
comparison_df = pd.DataFrame({
    "Actual_class": y_test.values,
    "Predicted_class": y_pred,
    "Prob_Class_0": y_prob[:, 0],
    "Prob_Class_1": y_prob[:, 1],
    "Prob_Class_2": y_prob[:, 2],
})
comparison_df.head()

# i.e for the first row, the actual class is 0, predicted class is 0 with a probability of 0.85 for class 0, 0.10 for class 1, and 0.05 for class 2.
```

# Model Evaluation

```python
# confusion matrix and classification report
from sklearn.metrics import confusion_matrix, classification_report
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# confusion matrix heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Pred_Class_0', 'Pred_Class_1', 'Pred_Class_2'],
            yticklabels=['Actual_Class_0', 'Actual_Class_1', 'Actual_Class_2'])
plt.title('Confusion Matrix Heatmap')
plt.ylabel('Actual Class')
plt.xlabel('Predicted Class')
plt.show()  
# strong performance on class 0
# moderate performance on class 2
# poor performance on class 1 due to high class imbalance
```

```python
# overalll accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Overall Accuracy:", accuracy)
# accuracy might not be the best metric due to class imbalance
```

```python
# model diagnostics and evaluation
# pseudo R
from sklearn.metrics import log_loss
logloss = log_loss(y_test, y_prob)
print("Log Loss (Pseudo R):", logloss)

# meaning the lower the log loss, the better the model's performance. the model shows moderate performance but struggles with the minority class.
# the model explains approximately 44% of the variance in the outcome variable, indicating moderate predictive power.
```

```python
# odds ratios
import numpy as np
odds_ratios = pd.DataFrame(
    {
        'Feature': X.columns,
        'Class 0 OR': np.exp(model.coef_[0]),
        'Class 1 OR': np.exp(model.coef_[1]),
        'Class 2 OR': np.exp(model.coef_[2]),
    }
)
odds_ratios

# i.e., for a one-unit increase in the feature, the odds of being in that class versus the reference class change by the odds ratio, holding all other features constant.
# An odds ratio greater than 1 indicates increased odds of being in that class, while an odds ratio less than 1 indicates decreased odds.
```
